```python
import pandas as pd
import re
import numpy as np
from minisom import MiniSom
from sklearn.preprocessing import StandardScaler

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ –¥–≤—É—Ö —Ñ–∞–π–ª–æ–≤
df = pd.concat([pd.read_excel('1.xls'), pd.read_excel('2.xlsx')], ignore_index=True)

# –û—á–∏—Å—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df = df[['–ê–≤—Ç–æ—Ä', '–°–æ–æ–±—â–µ–Ω–∏–µ']].dropna(subset=['–ê–≤—Ç–æ—Ä', '–°–æ–æ–±—â–µ–Ω–∏–µ'])
df['–°–æ–æ–±—â–µ–Ω–∏–µ'] = df['–°–æ–æ–±—â–µ–Ω–∏–µ'].astype(str)

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ —á–∞—Å—Ç–æ—Ç—ã —Å–∏–º–≤–æ–ª–æ–≤
def count_char_frequency(text, char):
    text = str(text).lower()
    return text.count(char.lower()) / len(text) if len(text) > 0 else 0

# –°–æ–∑–¥–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
features = df.groupby('–ê–≤—Ç–æ—Ä').agg(
    –ú—Ä=('–°–æ–æ–±—â–µ–Ω–∏–µ', 'size'),
    Lp=('–°–æ–æ–±—â–µ–Ω–∏–µ', lambda x: x.str.len().mean())
).reset_index()

chars_to_count = {'—É': 'f—É', '—ç': 'f—ç', '–∞': 'f–∞', 
                 '—é': 'f—é', '–µ': 'f–µ', '—è': 'f—è', 
                 '–∏': 'f–∏', '!': 'f!', '–æ': 'f–æ', '?': 'f?'}

for char, col_name in chars_to_count.items():
    features[col_name] = df.groupby('–ê–≤—Ç–æ—Ä')['–°–æ–æ–±—â–µ–Ω–∏–µ'].apply(
        lambda x: count_char_frequency(x.str.cat(sep=' '), char)
    ).values

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features.iloc[:, 1:])

# –ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è SOM
som = MiniSom(5, 5, scaled_features.shape[1], sigma=0.5, learning_rate=0.5)
som.train_random(scaled_features, 1000)

# –ù–∞–∑–Ω–∞—á–µ–Ω–∏–µ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤
features['–ö–ª–∞—Å—Ç–µ—Ä'] = [som.winner(x)[0]*5 + som.winner(x)[1] for x in scaled_features]

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
features.to_excel('—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã_–∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏.xlsx', index=False)

# –í—ã–≤–æ–¥ –ø–µ—Ä–≤—ã—Ö 5 —Å—Ç—Ä–æ–∫
print(features.to_markdown(index=False))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = features

# –í–µ—Å–∞ –∫—Ä–∏—Ç–µ—Ä–∏–µ–≤ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ "—Ç—Ä–æ–ª–ª—å-–∏–Ω–¥–µ–∫—Å–∞"
WEIGHTS = {
    '–ú—Ä': 0.3,     # –ê–∫—Ç–∏–≤–Ω–æ—Å—Ç—å
    'f!': 0.25,    # –í–æ—Å–∫–ª–∏—Ü–∞–Ω–∏—è 
    'f?': 0.25,    # –í–æ–ø—Ä–æ—Å—ã
    'f—é': 0.1,     # –†–µ–¥–∫–∏–µ —Å–∏–º–≤–æ–ª—ã
    'f—ç': 0.1,     # –†–µ–¥–∫–∏–µ —Å–∏–º–≤–æ–ª—ã
    'Lp': 0.2      # –î–ª–∏–Ω–∞ —Å–æ–æ–±—â–µ–Ω–∏–π
}

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
scaler = StandardScaler()
scaled = pd.DataFrame(
    scaler.fit_transform(df[list(WEIGHTS.keys())]),
    columns=WEIGHTS.keys()
)

# –†–∞—Å—á–µ—Ç —Ç—Ä–æ–ª–ª—å-–∏–Ω–¥–µ–∫—Å–∞
df['–¢—Ä–æ–ª–ª—å-–∏–Ω–¥–µ–∫—Å'] = sum(scaled[col] * weight for col, weight in WEIGHTS.items())

# –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è
conditions = [
    df['–¢—Ä–æ–ª–ª—å-–∏–Ω–¥–µ–∫—Å'] > 1.5,
    df['–¢—Ä–æ–ª–ª—å-–∏–Ω–¥–µ–∫—Å'] > 0.5,
]
choices = ['üî¥ –í—ã—Å–æ–∫–∞—è', 'üü° –°—Ä–µ–¥–Ω—è—è']
df['–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å'] = np.select(conditions, choices, default='üü¢ –ù–∏–∑–∫–∞—è')

# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
result = df.sort_values('–¢—Ä–æ–ª–ª—å-–∏–Ω–¥–µ–∫—Å', ascending=False).round(3)
print(result[['–ê–≤—Ç–æ—Ä', '–ú—Ä', 'Lp', 'f!', 'f?', '–¢—Ä–æ–ª–ª—å-–∏–Ω–¥–µ–∫—Å', '–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å']].to_markdown(index=False))
```
