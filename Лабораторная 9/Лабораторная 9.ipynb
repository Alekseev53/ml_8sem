```python
import pandas as pd
import re
import numpy as np
from minisom import MiniSom
from sklearn.preprocessing import StandardScaler

# Объединение данных из двух файлов
df = pd.concat([pd.read_excel('1.xls'), pd.read_excel('2.xlsx')], ignore_index=True)

# Очистка данных
df = df[['Автор', 'Сообщение']].dropna(subset=['Автор', 'Сообщение'])
df['Сообщение'] = df['Сообщение'].astype(str)

# Функция для подсчета частоты символов
def count_char_frequency(text, char):
    text = str(text).lower()
    return text.count(char.lower()) / len(text) if len(text) > 0 else 0

# Создание таблицы характеристик
features = df.groupby('Автор').agg(
    Мр=('Сообщение', 'size'),
    Lp=('Сообщение', lambda x: x.str.len().mean())
).reset_index()

chars_to_count = {'у': 'fу', 'э': 'fэ', 'а': 'fа', 
                 'ю': 'fю', 'е': 'fе', 'я': 'fя', 
                 'и': 'fи', '!': 'f!', 'о': 'fо', '?': 'f?'}

for char, col_name in chars_to_count.items():
    features[col_name] = df.groupby('Автор')['Сообщение'].apply(
        lambda x: count_char_frequency(x.str.cat(sep=' '), char)
    ).values

# Нормализация данных
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features.iloc[:, 1:])

# Кластеризация SOM
som = MiniSom(5, 5, scaled_features.shape[1], sigma=0.5, learning_rate=0.5)
som.train_random(scaled_features, 1000)

# Назначение кластеров
features['Кластер'] = [som.winner(x)[0]*5 + som.winner(x)[1] for x in scaled_features]

# Сохранение результатов
features.to_excel('результаты_кластеризации.xlsx', index=False)

# Вывод первых 5 строк
print(features.to_markdown(index=False))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = features

# Веса критериев для расчета "тролль-индекса"
WEIGHTS = {
    'Мр': 0.3,     # Активность
    'f!': 0.25,    # Восклицания 
    'f?': 0.25,    # Вопросы
    'fю': 0.1,     # Редкие символы
    'fэ': 0.1,     # Редкие символы
    'Lp': 0.2      # Длина сообщений
}

# Нормализация данных
scaler = StandardScaler()
scaled = pd.DataFrame(
    scaler.fit_transform(df[list(WEIGHTS.keys())]),
    columns=WEIGHTS.keys()
)

# Расчет тролль-индекса
df['Тролль-индекс'] = sum(scaled[col] * weight for col, weight in WEIGHTS.items())

# Классификация
conditions = [
    df['Тролль-индекс'] > 1.5,
    df['Тролль-индекс'] > 0.5,
]
choices = ['🔴 Высокая', '🟡 Средняя']
df['Вероятность'] = np.select(conditions, choices, default='🟢 Низкая')

# Сортировка и форматирование
result = df.sort_values('Тролль-индекс', ascending=False).round(3)
print(result[['Автор', 'Мр', 'Lp', 'f!', 'f?', 'Тролль-индекс', 'Вероятность']].to_markdown(index=False))
```
